## Also use the appropriate methods to remove;

```python
#Wrong data & format
# Removes rows with age that is not an int
df2 = df2[pd.to_numeric(df2['age'], errors='coerce').notnull()] 
df2

```

```python
# Removes rows with a py-score that is not an int
df2 = df2[pd.to_numeric(df2['py-score'], errors='coerce').notnull()] 
df2
```

```python
#Remove duplicates
df2.drop_duplicates()
```

```python
# Perform some correlation
df2.corr(method='spearman')
```
