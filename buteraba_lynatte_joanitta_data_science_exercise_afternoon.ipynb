## SERIES

```python
#1.	Write a Pandas program to compare the elements of two Pandas series. 
#Sample Series [4,65,436,3,9], [7,0,3,897,9]
import pandas as pd

# Create the two sample Series
series1 = pd.Series([4, 65, 436, 3, 9])
series2 = pd.Series([7, 0, 3, 897, 9])

# Compare the elements of the two Series
comparison = series1 == series2

# Display the comparison result
print(comparison)
```

```python
#2.	Write a Pandas program to add, subtract, multiply and divide two Pandas series. 
#Sample Series: [2,4,6,8,14], [1,3,5,7,9]
import pandas as pd

# Create the two sample Series
series1 = pd.Series([2, 4, 6, 8, 14])
series2 = pd.Series([1, 3, 5, 7, 9])

# Addition
addition = series1 + series2
print("Addition:")
print(addition)

# Subtraction
subtraction = series1 - series2
print("\nSubtraction:")
print(subtraction)

# Multiplication
multiplication = series1 * series2
print("\nMultiplication:")
print(multiplication)

# Division
division = series1 / series2
print("\nDivision:")
print(division)
```

```python
#3.	Write a Pandas program to convert a dictionary to a Pandas series. 
#Sample dictionary: dictionary1 = {‘Josh’: 24, ‘Sam’: 36, ‘Peace’: 19, ‘Charles’: 65, ‘Tom’: 44}
# Sample dictionary
dictionary1 = {'Josh': 24, 'Sam': 36, 'Peace': 19, 'Charles': 65, 'Tom': 44}

# Convert the dictionary to a Pandas Series
series1 = pd.Series(dictionary1)

# Display the resulting Series
print(series1)
```

```python
#4.	Write a Pandas program to convert a given series to an array.
#Sample series: [‘Love’, 800, ‘Joy’, 789.9, ‘Peace’, True]
import numpy as np

# Sample Series
series1 = pd.Series(['Love', 800, 'Joy', 789.9, 'Peace', True])

# Convert the Series to an array
array1 = np.array(series1)

# Display the resulting array
print(array1)
```

```python
#5.	Write a Pandas program to display the most frequent value in the given series 
#and replace everything else as ‘Other’ in the series. (Use the ‘HomeTeamGoals’ column in the dataset provided)
import pandas as pd
df = pd.read_csv('AfricaCupofNationsMatches.csv')
df
# Get the most frequent value
most_frequent_value = df['HomeTeamGoals'].value_counts().idxmax()

# Replace everything else with 'Other'
df['HomeTeamGoals'] = df['HomeTeamGoals'].map(lambda x: most_frequent_value if x == most_frequent_value else 'Other')

# Print the updated DataFrame
print(df)




```

```python
#12.Write a Pandas program to add 2 rows to your DataFrame.

# Create new rows as dictionaries
new_rows = [{'HomeTeam': 'Kenya', 'AwayTeam': 'DRCongo', 'HomeTeamGoals': 2, 'AwayTeamGoals': 1},
            {'HomeTeam': 'Uganda', 'AwayTeam': 'Sudan', 'HomeTeamGoals': 3, 'AwayTeamGoals': 2}]
# Convert new_rows to DataFrame
new_rows_df = pd.DataFrame(new_rows)

# Append the new rows to the DataFrame
df = pd.concat([df, new_rows_df], ignore_index=True)

# Print the updated DataFrame
print(df)
```

```python
#13.Write a Pandas program to change the country ‘Uganda’ to ‘China’ in the ‘AwayTeam’ column of the DataFrame

# Remove leading/trailing whitespaces and replace 'Uganda' with 'China' in the 'AwayTeam' column
df['AwayTeam'] = df['AwayTeam'].str.strip().replace('Uganda', 'China')

# Print the updated DataFrame
print(df)
```
